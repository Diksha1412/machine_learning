```python
import ast

def convert(text):
    if isinstance(text, list):
        # already parsed
        return ' '.join(i['name'] for i in text if 'name' in i)

    if isinstance(text, str):
        try:
            items = ast.literal_eval(text)
            return ' '.join(i['name'] for i in items if 'name' in i)
        except:
            return ''

    return ''
```
